```python
# libraries used
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
import csv
```

```python
# Storing te URL which we want to scrap
my_url = 'https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=graphics%20card'
my_url
```

```python
# opening the page 
uclient = ureq(my_url)
```

```python
# grabbing the page and closing
page_html = uclient.read()
uclient.close()
```

```python
# HTML Parser
page_soup = soup(page_html,"html.parser")
```

```python
# format the HTML
print(page_soup.prettify())
```

```python
# To find all the class which has item container and store it in a variable and open the csv file
containers = page_soup.findAll("div",{"class":"item-container"})
container = containers[0]
with open('gcard.csv', 'w', newline="") as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["brand","title","shipping"])
```

```python
# For loop to get all the item container details and storing in the csv file
data = []
for container in containers:
    brand = container.findAll("div",{"class":"item-branding"})
    brand_name = brand[0].a.img["title"]
    #print(brand_name)
    title = container.findAll("a",{"class":"item-title"})
    title_name = title[0].text
    #print(title_name)
    shipping = container.findAll("li",{"class":"price-ship"})
    shipping_price = shipping[0].text.strip()
    #print(shipping_price)
    itemprice = container.findAll("li",{"class":"price-current"})
    pr = itemprice[0].strong.text
    ice = itemprice[0].sup.text
    price = pr+ice
    #print(price)
    data.append((brand_name,title_name,shipping_price, price))
    
    print(brand_name)
    print(title_name)
    print(shipping_price)
    print(price)
with open('gcard.csv', 'a', newline="") as csv_file:
        writer = csv.writer(csv_file)
        for brand_name, title_name, shipping_price, price in data:
            writer.writerow([brand_name, title_name, shipping_price, price])
```
